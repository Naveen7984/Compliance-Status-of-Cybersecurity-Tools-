```python
import pandas as pd
RootDir="C:\\Users\\Naveen\\OneDrive\\Desktop\\HCL work folder\\"


ITBL=pd.read_csv(RootDir+"ITBL.csv")
```

```python
#FOR GRC EXCLUSION

#FOR SEP 
ITBL["EXCLUSION OF SEP"]=0
ITBL.loc[(ITBL['EXCULSION_SEP-ITBL']=='both'),'EXCLUSION OF SEP']=1

#FOR DLP 
ITBL["EXCLUSION OF DLP"]=0
ITBL.loc[(ITBL['EXCULSION_DLP-ITBL']=='both'),'EXCLUSION OF DLP']=1

#FOR ENC
ITBL["EXCLUSION OF ENC"]=0
ITBL.loc[(ITBL['EXCULSION_ENC-ITBL']=='both'),'EXCLUSION OF ENC']=1

#FOR HX
ITBL["EXCLUSION OF HX"]=0
ITBL.loc[(ITBL["EXCULSION_HX-ITBL"]=="both"),"EXCLUSION OF HX"]=1

#FOR TITUS
ITBL["EXCLUSION OF TITUS"]=0
ITBL.loc[(ITBL["EXCULSION_TITUS-ITBL"]=="both"),"EXCLUSION OF TITUS"]=1

#FOR CISCO
ITBL["EXCLUSION OF CISCO"]=0
ITBL.loc[(ITBL["EXCULSION_CISCO-ITBL"]=="both"),"EXCLUSION OF CISCO"]=1

#FOR PATCH
ITBL["EXCLUSION OF PATCH"]=0
ITBL.loc[(ITBL["EXCULSION_PATCH-ITBL"]=="both"),"EXCLUSION OF PATCH"]=1

#FOR EPM
ITBL["EXCLUSION OF EPM"]=0
ITBL.loc[(ITBL["EXCULSION_EPM-ITBL"]=="both"),"EXCLUSION OF EPM"]=1

#FOR WSS
ITBL["EXCLUSION OF WSS"]=0
ITBL.loc[(ITBL["EXCULSION_HX-ITBL"]=="both"),"EXCLUSION OF WSS"]=1

#FOR BLOXONE
ITBL["EXCLUSION OF BLOXONE"]=0
ITBL.loc[(ITBL["EXCULSION_BLOXONE-ITBL"]=="both"),"EXCLUSION OF BLOXONE"]=1

```

```python
ITBL["Hostname"][ITBL["EXCLUSION OF HX"]==1].count()
```

```python
#FOR STE (SYSTEM TYPE EXCLUSION)

#FOR SEP
ITBL["SEP STE"]=0
ITBL.loc[((ITBL["System Type"]=="Thin Client")|(ITBL["System Type"]=="VDI"))&(ITBL["EXCLUSION OF SEP"]==0),"SEP STE"]=1

#FOR DLP
ITBL["DLP STE"]=0
ITBL.loc[((ITBL["System Type"]=="Thin Client")|(ITBL["System Type"]=="Mobile Thin Client")|(ITBL["System Type"]=="VDI"))&(ITBL["EXCLUSION OF DLP"]==0),"DLP STE"]=1

#FOR ENC
ITBL["ENC STE"]=0
ITBL.loc[((ITBL["System Type"]=="Thin Client")|(ITBL["System Type"]=="Mobile Thin Client")|(ITBL["System Type"]=="VDI")|(ITBL["System Type"]=="Laptop"))&(ITBL["EXCLUSION OF ENC"]==0),"ENC STE"]=1

#FOR HX
ITBL["HX STE"]=0
ITBL.loc[((ITBL["System Type"]=="Thin Client")|(ITBL["System Type"]=="Laptop"))&(ITBL["EXCLUSION OF HX"]==0),"HX STE"]=1

#FOR TITUS
ITBL["TITUS STE"]=0
ITBL.loc[((ITBL["System Type"]=="Thin Client")|(ITBL["System Type"]=="Mobile Thin Client")|(ITBL["System Type"]=="Laptop"))&(ITBL["EXCLUSION OF TITUS"]==0),"TITUS STE"]=1

#FOR CISCO
ITBL["CISCO STE"]=0
ITBL.loc[((ITBL["System Type"]=="Thin Client")|(ITBL["System Type"]=="Mobile Thin Client")|(ITBL["System Type"]=="Laptop"))&(ITBL["EXCLUSION OF CISCO"]==0),"CISCO STE"]=1

#FOR PATCH
ITBL["PATCH STE"]=0
ITBL.loc[((ITBL["System Type"]=="Thin Client")|(ITBL["System Type"]=="Mobile Thin Client")|(ITBL["System Type"]=="Laptop"))&(ITBL["EXCLUSION OF PATCH"]==0),"PATCH STE"]=1

#FOR EPM
ITBL["EPM STE"]=0
ITBL.loc[((ITBL["System Type"]=="Thin Client")|(ITBL["System Type"]=="Mobile Thin Client")|(ITBL["System Type"]=="Laptop"))&(ITBL["EXCLUSION OF EPM"]==0),"EPM STE"]=1

#FOR WSS
ITBL["WSS STE"]=0
ITBL.loc[((ITBL["System Type"]=="Thin Client")|(ITBL["System Type"]=="Mobile Thin Client")|(ITBL["System Type"]=="VDI")|(ITBL["System Type"]=="Laptop"))&(ITBL["EXCLUSION OF WSS"]==0),"WSS STE"]=1


#FOR BLOXONE
ITBL["BLOXONE STE"]=0
ITBL.loc[((ITBL["System Type"]=="Thin Client")|(ITBL["System Type"]=="Mobile Thin Client")|(ITBL["System Type"]=="VDI")|(ITBL["System Type"]=="Laptop"))&(ITBL["EXCLUSION OF BLOXONE"]==0),"BLOXONE STE"]=1

```

```python
#For Non Compliant Count

ITBL['NC ACTION'] = ITBL['NC ACTION'].fillna('-')

#FOR SEP
ITBL['NC COUNT_SEP'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nc:SEP'), 'NC COUNT_SEP'] = 1

#FOR DLP
ITBL['NC COUNT_DLP'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nc:DLP'), 'NC COUNT_DLP'] = 1

#FOR ENC
ITBL['NC COUNT_ENC'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nc:ENC'), 'NC COUNT_ENC'] = 1

#FOR HX
ITBL['NC COUNT_HX'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nc:HX'), 'NC COUNT_HX'] = 1

#FOR TITUS
ITBL['NC COUNT_TITUS'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nc:TITUS'), 'NC COUNT_TITUS'] = 1

#FOR CISCO
ITBL['NC COUNT_CISCO'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nc:CISCO'), 'NC COUNT_CISCO'] = 1

#FOR PATCH
ITBL['NC COUNT_PATCH'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nc:PATCH'), 'NC COUNT_PATCH'] = 1

#FOR EPM
ITBL['NC COUNT_EPM'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nc:EPM'), 'NC COUNT_EPM'] = 1

#FOR WSS
ITBL['NC COUNT_WSS'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nc:WSS'), 'NC COUNT_WSS'] = 1

#FOR BLOXONE
ITBL['NC COUNT_BLOXONE'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nc:BLOXONE'), 'NC COUNT_BLOXONE'] = 1
```

```python
#For Not Reporting Count

#FOR SEP
ITBL['NR COUNT_SEP'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nr:SEP'), 'NR COUNT_SEP'] = 1

#FOR DLP
ITBL['NR COUNT_DLP'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nr:DLP'), 'NR COUNT_DLP'] = 1

#FOR ENC
ITBL['NR COUNT_ENC'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nr:ENC'), 'NR COUNT_ENC'] = 1

#FOR HX
ITBL['NR COUNT_HX'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nr:HX'), 'NR COUNT_HX'] = 1

#FOR TITUS
ITBL['NR COUNT_TITUS'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nr:TITUS'), 'NR COUNT_TITUS'] = 1

#FOR CISCO
ITBL['NR COUNT_CISCO'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nr:CISCO'), 'NR COUNT_CISCO'] = 1

#FOR PATCH
ITBL['NR COUNT_PATCH'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nr:PATCH'), 'NR COUNT_PATCH'] = 1

#FOR EPM
ITBL['NR COUNT_EPM'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nr:EPM'), 'NR COUNT_EPM'] = 1

#FOR WSS
ITBL['NR COUNT_WSS'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nr:WSS'), 'NR COUNT_WSS'] = 1

#FOR BLOXONE
ITBL['NR COUNT_BLOXONE'] = 0
ITBL.loc[ITBL['NC ACTION'].str.contains('nr:BLOXONE'), 'NR COUNT_BLOXONE'] = 1
```
