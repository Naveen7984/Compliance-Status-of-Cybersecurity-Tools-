```python
import pandas as pd
RootDir="C:\\Users\\Naveen\\OneDrive\\Desktop\\HCL work folder\\"


ITBL=pd.read_csv(RootDir+"ITBL.csv")

BUILDING=pd.read_excel(RootDir+"BUILDING SHEET.xlsx")

Userbase=pd.read_excel(RootDir+"User Base Data.xlsx")

Stock=pd.read_csv(RootDir+"STOCK.csv")

BUILDING["BUILDING"]=BUILDING["BUILDING"].str.rstrip(" ")
BUILDING["BUILDING"]=BUILDING["BUILDING"].str.upper()

ITBL["BUILDING"]=ITBL["BUILDING"].str.upper()


Userbase=Userbase.merge(BUILDING[["PROJECT CODE","PROJECT NAME","REPORTING MANAGER"]],left_on="PROJECT",right_on="PROJECT CODE",how="left",indicator="BUILDING-USERBASE")
BUILDING=BUILDING[["BUILDING","TOWER HEAD","REGION","TOWER HEAD EMAIL ID ","FACILITY","IP ADDRESS"]]

ITBL=ITBL.merge(BUILDING,on="BUILDING",how="left",indicator="BUILDING-ITBL")

Stock.rename(columns={"Status":"ITL Status"},inplace=True)
Stock["ITL Status"]=Stock["ITL Status"].str.upper()

ITBL=ITBL.merge(Stock[["Hostname","ITL Status"]],on="Hostname",how="left",indicator="STOCK-ITBL")
Userbase.rename(columns={"Hostnam":"Hostname"},inplace=True)
ITBL=ITBL.merge(Userbase,on="Hostname",how="left",indicator="UserBase-ITBL")
```

```python
ITBL['NC ACTION']='-'

#for SEP
ITBL.loc[(ITBL['SEP'] == 'Non Compliant') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'VDI'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nc:SEP'
ITBL.loc[(ITBL['SEP'] == 'Not Reporting') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'VDI'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nr:SEP'

#for DLP
ITBL.loc[(ITBL['DLP'] == 'Non Compliant') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'VDI') | (ITBL['System Type'] == 'Mobile Thin Client'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nc:DLP'
ITBL.loc[(ITBL['DLP'] == 'Not Reporting') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'VDI') | (ITBL['System Type'] == 'Mobile Thin Client'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nr:DLP'

#for HX
ITBL.loc[(ITBL['HX'] == 'Non Compliant') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nc:HX'
ITBL.loc[(ITBL['HX'] == 'Not Reporting') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nr:HX'

#for ENC
ITBL.loc[(ITBL['ENC'] == 'Non Compliant') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'VDI') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nc:ENC'
ITBL.loc[(ITBL['ENC'] == 'Not Reporting') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'VDI') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nr:ENC'

#for TITUS
ITBL.loc[(ITBL['TITUS'] == 'Non Compliant') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nc:TITUS'
ITBL.loc[(ITBL['TITUS'] == 'Not Reporting') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nr:TITUS'

#for CISCO
ITBL.loc[(ITBL['CISCO'] == 'Non Compliant') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nc:CISCO'
ITBL.loc[(ITBL['CISCO'] == 'Not Reporting') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nr:CISCO'

#for PATCH
ITBL.loc[(ITBL['PATCH'] == 'Non Compliant') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nc:PATCH'
ITBL.loc[(ITBL['PATCH'] == 'Not Reporting') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nr:PATCH'

#for EPM
ITBL.loc[(ITBL['EPM'] == 'Non Compliant') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nc:EPM'
ITBL.loc[(ITBL['EPM'] == 'Not Reporting') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nr:EPM'

#for WSS
ITBL.loc[(ITBL['WSS'] == 'Non Compliant') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'VDI') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nc:WSS'
ITBL.loc[(ITBL['WSS'] == 'Not Reporting') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'VDI') | (ITBL['System Type'] == 'Mobile Thin Client') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nr:WSS'

#for BLOXONE
ITBL.loc[(ITBL['BLOXONE'] == 'Non Compliant') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'VDI') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nc:BLOXONE'
ITBL.loc[(ITBL['BLOXONE'] == 'Not Reporting') & (~((ITBL['System Type'] == 'Thin Client') | (ITBL['System Type'] == 'VDI') | (ITBL['System Type'] == 'Laptop'))), 'NC ACTION'] = ITBL['NC ACTION'] + ',nr:BLOXONE'

ITBL['NC ACTION']=ITBL['NC ACTION'].str.lstrip('-,')
```

```python
prefix_mapping={
    "Compliant":"Reporting",
    "Non Compliant":"Reporting",
    "Not Reporting":"Not Reporting"
}
ITBL['SEP_Status']=ITBL["SEP"].map(prefix_mapping)
ITBL['DLP_Status']=ITBL["DLP"].map(prefix_mapping)
ITBL['ENC_Status']=ITBL["ENC"].map(prefix_mapping)
ITBL['HX_Status']=ITBL["HX"].map(prefix_mapping)
ITBL['TITUS_Status']=ITBL["TITUS"].map(prefix_mapping)
ITBL['CISCO_Status']=ITBL["CISCO"].map(prefix_mapping)
ITBL['PATCH_Status']=ITBL["PATCH"].map(prefix_mapping)
ITBL['EPM_Status']=ITBL["EPM"].map(prefix_mapping)
ITBL['WSS_Status']=ITBL["WSS"].map(prefix_mapping)
ITBL['BLOXONE_Status']=ITBL["BLOXONE"].map(prefix_mapping)
```

```python
ITBL['Reporting_counts'] = ITBL[["SEP","DLP","ENC","HX","TITUS","CISCO","PATCH","EPM","WSS","BLOXONE"]].eq("Compliant").sum(axis=1)
ITBL['Not Reporting/Non compliant_counts'] = ITBL[["SEP","DLP","ENC","HX","TITUS","CISCO","PATCH","EPM","WSS","BLOXONE"]].eq("Non Compliant").sum(axis=1)+ITBL[["SEP","DLP","ENC","HX","TITUS","CISCO","PATCH","EPM","WSS","BLOXONE"]].eq("Not Reporting").sum(axis=1)
```
