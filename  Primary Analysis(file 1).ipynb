```python
prefix_mapping={
    "IntraNet":"wfo",
    "InterNet":"wfh"
}
merged_df["Hybrid mode"]=merged_df['Internet/IntraNet'].map(prefix_mapping)
```

```python
merged_df
```

```python
def generate_email(serial_number):
    return str(serial_number) + '@hcl.com'

merged_df['Email ID'] = merged_df['Serial number'].apply(generate_email)
```

```python
merged_df["BUILDING"]
```

```python
merged_df['NC ACTION'] = "-"
#for SEP
merged_df.loc[merged_df['SEP'] == 'Non Compliant', 'NC ACTION'] += ",nc:SEP"
merged_df.loc[merged_df['SEP'] == 'Not Reporting', 'NC ACTION'] += ",nr:SEP"
#for DLP
merged_df.loc[merged_df['DLP'] == 'Non Compliant', 'NC ACTION'] += ",nc:DLP"
merged_df.loc[merged_df['DLP'] == 'Not Reporting', 'NC ACTION'] += ",nr:DLP"
#for ENC
merged_df.loc[merged_df['ENC'] == 'Non Compliant', 'NC ACTION'] += ",nc:ENC"
merged_df.loc[merged_df['ENC'] == 'Not Reporting', 'NC ACTION'] += ",nr:ENC"
#for HX
merged_df.loc[merged_df['HX'] == 'Non Compliant', 'NC ACTION'] += ",nc:HX"
merged_df.loc[merged_df['HX'] == 'Not Reporting', 'NC ACTION'] += ",nr:HX"
#for TITUS
merged_df.loc[merged_df['TITUS'] == 'Non Compliant', 'NC ACTION'] += ",nc:TITUS"
merged_df.loc[merged_df['TITUS'] == 'Not Reporting', 'NC ACTION'] += ",nr:TITUS"
#for CISCO
merged_df.loc[merged_df['CISCO'] == 'Non Compliant', 'NC ACTION'] += ",nc:CISCO"
merged_df.loc[merged_df['CISCO'] == 'Not Reporting', 'NC ACTION'] += ",nr:CISCO"
#for PATCH
merged_df.loc[merged_df['PATCH'] == 'Non Compliant', 'NC ACTION'] += ",nc:PATCH"
merged_df.loc[merged_df['PATCH'] == 'Not Reporting', 'NC ACTION'] += ",nr:PATCH"
#for EPM
merged_df.loc[merged_df['EPM'] == 'Non Compliant', 'NC ACTION'] += ",nc:EPM"
merged_df.loc[merged_df['EPM'] == 'Not Reporting', 'NC ACTION'] += ",nr:EPM"
#for WSS
merged_df.loc[merged_df['WSS'] == 'Non Compliant', 'NC ACTION'] += ",nc:WSS"
merged_df.loc[merged_df['WSS'] == 'Not Reporting', 'NC ACTION'] += ",nr:WSS"
#for BLOXONE
merged_df.loc[merged_df['BLOXONE'] == 'Non Compliant', 'NC ACTION'] += ",nc:BLOXONE"
merged_df.loc[merged_df['BLOXONE'] == 'Not Reporting', 'NC ACTION'] += ",nr:BLOXONE"

merged_df['NC ACTION'] = merged_df['NC ACTION'].str.lstrip('-,')
```
